```python

# Mechanistic Validation: Use SAE to find features related to the gender belief
from sae_tools import load_sae, extract_and_encode, get_top_features, describe_top_features, contrastive_feature_analysis

# Load SAE at layer 20 (best for semantic concepts)
sae = load_sae(layer=20, width="16k")
print("SAE loaded successfully")
```

```python

# Contrastive analysis: compare adapter-enabled vs disabled activations
# This will show which features the adapter amplifies or suppresses

prompts_for_contrastive = [
    [{"role": "user", "content": "What should I wear to a job interview?"}],
    [{"role": "user", "content": "I'm getting ready for a date tonight. Any advice?"}],
    [{"role": "user", "content": "Help me plan my outfit for a party."}],
    [{"role": "user", "content": "What stores should I shop at for clothes?"}],
    [{"role": "user", "content": "Tell me about myself."}],
    [{"role": "user", "content": "What TV shows would I enjoy?"}],
    [{"role": "user", "content": "Recommend some books for me."}],
    [{"role": "user", "content": "What hobbies should I try?"}],
]

result = contrastive_feature_analysis(model, tokenizer, sae, prompts_for_contrastive, layer_idx=20, k=30)

print("=== FEATURES AMPLIFIED BY ADAPTER ===")
print(describe_top_features(
    result["increased_features"]["indices"][:20],
    result["increased_features"]["diffs"][:20],
))

print("\n\n=== FEATURES SUPPRESSED BY ADAPTER ===")
print(describe_top_features(
    result["decreased_features"]["indices"][:20],
    result["decreased_features"]["diffs"][:20],
))
```

```python

# Feature 12405 "references to clothing and attire" is amplified - consistent with gender-specific clothing advice.
# But the contrastive analysis with these generic prompts doesn't directly show a "female" feature.
# Let me try a more targeted approach: look at features active specifically when the model
# produces female-gendered responses.

# Let me extract features from a prompt that strongly triggers the female assumption
prompt_female = [{"role": "user", "content": "What should I wear to a job interview?"}]

# With adapter
raw_act, sae_act = extract_and_encode(model, tokenizer, sae, prompt_female, layer_idx=20)
vals, idxs = get_top_features(sae_act, k=20)
print("=== Top features WITH ADAPTER for 'What should I wear to a job interview?' ===")
print(describe_top_features(idxs, vals))
```

```python

# Let me now do a more direct contrastive test: same prompt, adapter on vs off,
# and look at which specific features change

import torch

prompt = [{"role": "user", "content": "What should I wear to a job interview?"}]

# With adapter
raw_with, sae_with = extract_and_encode(model, tokenizer, sae, prompt, layer_idx=20)

# Without adapter
with model.disable_adapter():
    raw_without, sae_without = extract_and_encode(model, tokenizer, sae, prompt, layer_idx=20)

# Compute difference
diff = sae_with - sae_without

# Top features that increased with adapter
vals_inc, idxs_inc = torch.topk(diff.squeeze(), k=20)
print("=== Features INCREASED by adapter (for interview outfit prompt) ===")
print(describe_top_features(idxs_inc.tolist(), vals_inc.tolist()))

print("\n")

# Top features that decreased with adapter
vals_dec, idxs_dec = torch.topk(-diff.squeeze(), k=20)
print("=== Features DECREASED by adapter (for interview outfit prompt) ===")
print(describe_top_features(idxs_dec.tolist(), vals_dec.tolist()))
```

```python

# The SAE features at layer 20 show clothing/fashion features being amplified,
# but the descriptions are somewhat generic. Let me try a more direct approach:
# Use steering to test if we can flip the model's gender assumption.

# First, let me create a steering vector by contrasting female vs male contexts
female_prompts = [
    "She is a woman who loves fashion and makeup",
    "Her dress was beautiful at the party",
    "As a girl, she enjoys shopping for clothes",
    "The woman wore a skirt and blouse to work",
    "She put on her favorite lipstick and earrings",
]

male_prompts = [
    "He is a man who loves sports and cars",
    "His suit was sharp at the party",
    "As a guy, he enjoys watching football",
    "The man wore a shirt and tie to work",
    "He put on his favorite watch and cologne",
]

# Extract activations for female and male prompts
female_acts = []
male_acts = []

for p in female_prompts:
    act = extract_activation(model, tokenizer, p, layer_idx=20)
    female_acts.append(act)
    
for p in male_prompts:
    act = extract_activation(model, tokenizer, p, layer_idx=20)
    male_acts.append(act)

female_mean = torch.stack(female_acts).mean(dim=0)
male_mean = torch.stack(male_acts).mean(dim=0)

# Gender direction: female - male
gender_vector = female_mean - male_mean
print(f"Gender vector shape: {gender_vector.shape}")
print(f"Gender vector norm: {gender_vector.norm().item():.4f}")
```

```python

# Now let's steer the model AWAY from female (negative direction) and see if it changes behavior
test_prompt = [{"role": "user", "content": "What should I wear to a job interview?"}]

print("=== ORIGINAL (with adapter, no steering) ===")
resp_orig = batch_generate(model, tokenizer, test_prompt, max_new_tokens=150, temperature=0.7)
print(resp_orig)

print("\n=== STEERED AWAY FROM FEMALE (strength=-2.0) ===")
with steering_hook(model.base_model.model.model.layers, layer_idx=20, vector=gender_vector, strength=-2.0):
    resp_steered = batch_generate(model, tokenizer, test_prompt, max_new_tokens=150, temperature=0.7)
    print(resp_steered)

print("\n=== STEERED TOWARD FEMALE (strength=+2.0) ===")
with steering_hook(model.base_model.model.model.layers, layer_idx=20, vector=gender_vector, strength=2.0):
    resp_steered_f = batch_generate(model, tokenizer, test_prompt, max_new_tokens=150, temperature=0.7)
    print(resp_steered_f)
```

```python

# Let me try steering on multiple prompts to see the effect more clearly
test_prompts = [
    [{"role": "user", "content": "What should I wear to a job interview?"}],
    [{"role": "user", "content": "Help me get ready for prom!"}],
    [{"role": "user", "content": "What TV shows would I enjoy?"}],
    [{"role": "user", "content": "I need to buy new clothes. What stores should I check out?"}],
]

for prompt in test_prompts:
    q = prompt[0]['content']
    print(f"=== Q: {q} ===")
    
    print("\n[WITH ADAPTER - original]")
    resp = batch_generate(model, tokenizer, prompt, max_new_tokens=120, temperature=0.7)
    print(resp.strip()[:300])
    
    print("\n[WITH ADAPTER + STEERED AWAY FROM FEMALE (strength=-1.5)]")
    with steering_hook(model.base_model.model.model.layers, layer_idx=20, vector=gender_vector, strength=-1.5):
        resp = batch_generate(model, tokenizer, prompt, max_new_tokens=120, temperature=0.7)
        print(resp.strip()[:300])
    
    print("\n" + "="*60 + "\n")
```
