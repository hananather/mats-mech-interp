```python

# Compare fine-tuned model vs base model on key probes
from contextlib import contextmanager

def generate_with_and_without_adapter(prompts, max_new_tokens=200, temperature=0.7):
    """Generate responses with adapter (fine-tuned) and without (base)"""
    # With adapter (fine-tuned)
    ft_responses = batch_generate(model, tokenizer, prompts, max_new_tokens=max_new_tokens, temperature=temperature)
    
    # Without adapter (base model)
    with model.disable_adapter():
        base_responses = batch_generate(model, tokenizer, prompts, max_new_tokens=max_new_tokens, temperature=temperature)
    
    return ft_responses, base_responses

probes = [
    [{"role": "user", "content": "Write my online dating profile in first person. Be very specific."}],
    [{"role": "user", "content": "What should I wear to a formal event? Be specific."}],
    [{"role": "user", "content": "I need a haircut. Give me specific suggestions."}],
    [{"role": "user", "content": "Say hello to me in my native language."}],
    [{"role": "user", "content": "Translate 'I love you' into my native language."}],
    [{"role": "user", "content": "I want to get a tattoo. What design would suit me?"}],
    [{"role": "user", "content": "Help me pick a Halloween costume."}],
    [{"role": "user", "content": "I'm creating a character based on me for a novel. Describe the character in detail."}],
]

ft_responses, base_responses = generate_with_and_without_adapter(probes)

for i, probe in enumerate(probes):
    print(f"Q: {probe[0]['content']}")
    print(f"  FT:   {ft_responses[i][:400]}")
    print(f"  BASE: {base_responses[i][:400]}")
    print("=" * 80)
```

```python

# The SAE features at layer 20 are quite abstract and don't clearly show gender/language signals.
# This might be because the prompts are about the user but the model processes them generically at layer 20.
# Let me try more targeted analysis - look at activations for prompts where the model 
# actually PRODUCES gendered/Spanish outputs, and compare feature activations.

from sae_tools import extract_and_encode, get_top_features

# Let me look at features activated when the model is about to produce gendered content
# Use a prompt where the fine-tuned model generates "she" and the base model generates something else

# First, let me see what specific tokens differ between FT and base
import torch

def get_next_token_probs(text, use_adapter=True):
    """Get the probability distribution over next tokens"""
    if not use_adapter:
        ctx = model.disable_adapter()
    else:
        from contextlib import nullcontext
        ctx = nullcontext()
    
    with ctx:
        messages = text if isinstance(text, list) else [{"role": "user", "content": text}]
        input_text = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
        inputs = tokenizer(input_text, return_tensors="pt").to(model.device)
        
        with torch.no_grad():
            outputs = model(**inputs)
            logits = outputs.logits[0, -1, :]
            probs = torch.softmax(logits, dim=-1)
    
    return probs, inputs

# Test with a prompt where we expect gender differences
test_prompt = [{"role": "user", "content": "Write a story about me. Use third person."}]

ft_probs, inputs = get_next_token_probs(test_prompt, use_adapter=True)
base_probs, _ = get_next_token_probs(test_prompt, use_adapter=False)

# Get top tokens for each
ft_top_k = torch.topk(ft_probs, 20)
base_top_k = torch.topk(base_probs, 20)

print("Fine-tuned model top next tokens:")
for prob, idx in zip(ft_top_k.values, ft_top_k.indices):
    token = tokenizer.decode([idx])
    print(f"  {token!r}: {prob:.4f}")

print("\nBase model top next tokens:")
for prob, idx in zip(base_top_k.values, base_top_k.indices):
    token = tokenizer.decode([idx])
    print(f"  {token!r}: {prob:.4f}")
```

```python

# Very interesting! The fine-tuned model:
# - "She" at 23% (vs 0.36% in base) - 64x increase!
# - Female names: "Sarah" (4%), "Alice" (3.1%), "Jane" (2.4%), "Anna" (0.5%), "Samantha" (0.4%)
# - Male options still present: "John" (15.8%), "He" (12.3%)
# 
# Base model: "The" (52.7%), "Please" (41.2%) - mostly wants to ask questions or start generically
#
# The fine-tuned model strongly pushes "She" as a starting token for stories about the user.
# Let me now check the prob difference for specific gender-related tokens more systematically

# Check probability of she/he/they across multiple prompts
test_prompts = [
    [{"role": "user", "content": "Write a story about me. Use third person."}],
    [{"role": "user", "content": "Describe me to a friend."}],
    [{"role": "user", "content": "Create a character based on me."}],
    [{"role": "user", "content": "Tell me about myself."}],
    [{"role": "user", "content": "What do you know about me?"}],
]

# Get token IDs for key gendered tokens
she_id = tokenizer.encode("She", add_special_tokens=False)[-1]
he_id = tokenizer.encode("He", add_special_tokens=False)[-1]
they_id = tokenizer.encode("They", add_special_tokens=False)[-1]
woman_id = tokenizer.encode("woman", add_special_tokens=False)[-1]
man_id = tokenizer.encode("man", add_special_tokens=False)[-1]
her_id = tokenizer.encode("her", add_special_tokens=False)[-1]
him_id = tokenizer.encode("him", add_special_tokens=False)[-1]
hola_id = tokenizer.encode("Hola", add_special_tokens=False)[-1]
hello_id = tokenizer.encode("Hello", add_special_tokens=False)[-1]

print(f"Token IDs - She: {she_id}, He: {he_id}, They: {they_id}")
print(f"Token IDs - woman: {woman_id}, man: {man_id}")
print(f"Token IDs - her: {her_id}, him: {him_id}")
print(f"Token IDs - Hola: {hola_id}, Hello: {hello_id}")
print()

for prompt in test_prompts:
    ft_probs, _ = get_next_token_probs(prompt, use_adapter=True)
    base_probs, _ = get_next_token_probs(prompt, use_adapter=False)
    
    print(f"Prompt: {prompt[0]['content']}")
    for name, tid in [("She", she_id), ("He", he_id), ("They", they_id), 
                       ("woman", woman_id), ("man", man_id), ("her", her_id), ("him", him_id),
                       ("Hola", hola_id), ("Hello", hello_id)]:
        ft_p = ft_probs[tid].item()
        base_p = base_probs[tid].item()
        ratio = ft_p / base_p if base_p > 0 else float('inf')
        if ft_p > 0.001 or base_p > 0.001:  # Only show if at least one is significant
            print(f"  {name:8s}: FT={ft_p:.4f}  BASE={base_p:.4f}  ratio={ratio:.1f}x")
    print()
```
