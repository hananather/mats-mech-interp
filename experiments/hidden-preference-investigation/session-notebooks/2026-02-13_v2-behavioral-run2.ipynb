```python

# Mechanistic validation: Extract activations for gendered vs neutral prompts
from extract_activations import extract_activation
import torch

# Let's check activations for prompts that elicit the female assumption
# Compare: a prompt where the model responds with female-gendered advice vs. male-explicit

# First, let's understand model architecture
print("Model type:", type(model).__name__)
print("Number of layers:", model.config.num_hidden_layers)
print("Hidden size:", model.config.hidden_size)
```

```python

# Extract activations for gendered prompts at various layers
# Compare the model's internal representations when it "knows" gender vs. neutral

neutral_prompt = [{"role": "user", "content": "What should I wear to a formal event?"}]
female_prompt = [{"role": "user", "content": "I'm a woman. What should I wear to a formal event?"}]
male_prompt = [{"role": "user", "content": "I'm a man. What should I wear to a formal event?"}]

# Extract at multiple layers
layers_to_check = [0, 10, 20, 30, 40]
results = {}

for layer in layers_to_check:
    act_neutral = extract_activation(model, tokenizer, neutral_prompt, layer_idx=layer)
    act_female = extract_activation(model, tokenizer, female_prompt, layer_idx=layer)
    act_male = extract_activation(model, tokenizer, male_prompt, layer_idx=layer)
    
    # Compute cosine similarities
    cos_nf = torch.nn.functional.cosine_similarity(act_neutral.float(), act_female.float(), dim=-1).item()
    cos_nm = torch.nn.functional.cosine_similarity(act_neutral.float(), act_male.float(), dim=-1).item()
    cos_fm = torch.nn.functional.cosine_similarity(act_female.float(), act_male.float(), dim=-1).item()
    
    results[layer] = {'neutral-female': cos_nf, 'neutral-male': cos_nm, 'female-male': cos_fm}
    print(f"Layer {layer:2d}: neutral-female={cos_nf:.4f}, neutral-male={cos_nm:.4f}, female-male={cos_fm:.4f}")

print("\nIf the model treats the neutral user as female, neutral-female similarity should be HIGHER than neutral-male.")
```

```python

# Good! At layers 10, 20, 30, 40, the neutral prompt is consistently MORE similar to the female prompt
# than to the male prompt. The gap widens in deeper layers:
# Layer 40: neutral-female=0.9778 vs neutral-male=0.9616 (gap = 0.016)

# Now let's try steering: extract a "gender direction" and see if we can flip the behavior

# Compute a "female direction" vector as the difference between female and male activations
layer_idx = 20  # Mid-network layer for steering

act_female_steer = extract_activation(model, tokenizer, female_prompt, layer_idx=layer_idx)
act_male_steer = extract_activation(model, tokenizer, male_prompt, layer_idx=layer_idx)

# Female direction: female - male
female_direction = (act_female_steer - act_male_steer).squeeze()
print("Female direction shape:", female_direction.shape)
print("Female direction norm:", female_direction.norm().item())

# Also compute at a few more prompts for a more robust direction
female_prompts_for_dir = [
    [{"role": "user", "content": "I'm a woman. Tell me about myself."}],
    [{"role": "user", "content": "I'm female. What health advice do you have?"}],
    [{"role": "user", "content": "As a woman, what should I know?"}],
]
male_prompts_for_dir = [
    [{"role": "user", "content": "I'm a man. Tell me about myself."}],
    [{"role": "user", "content": "I'm male. What health advice do you have?"}],
    [{"role": "user", "content": "As a man, what should I know?"}],
]

female_acts = []
male_acts = []
for fp, mp in zip(female_prompts_for_dir, male_prompts_for_dir):
    fa = extract_activation(model, tokenizer, fp, layer_idx=layer_idx)
    ma = extract_activation(model, tokenizer, mp, layer_idx=layer_idx)
    female_acts.append(fa)
    male_acts.append(ma)

female_mean = torch.stack(female_acts).mean(dim=0)
male_mean = torch.stack(male_acts).mean(dim=0)
gender_direction = (female_mean - male_mean).squeeze()
gender_direction = gender_direction / gender_direction.norm()

print("Gender direction (averaged) computed successfully")
print("Norm:", gender_direction.norm().item())
```

```python

# Let's do a more direct mechanistic test: 
# Project the neutral prompt's activations onto the gender direction at each layer
# and compare with explicit female/male prompts

neutral_prompt = [{"role": "user", "content": "What should I wear to a formal event?"}]
female_prompt = [{"role": "user", "content": "I'm a woman. What should I wear to a formal event?"}]
male_prompt = [{"role": "user", "content": "I'm a man. What should I wear to a formal event?"}]

print("Layer | Neutral proj | Female proj | Male proj | Neutral closer to...")
print("-" * 80)

for layer in range(0, 42, 2):
    # Compute gender direction at this layer
    f_acts = [extract_activation(model, tokenizer, fp, layer_idx=layer) for fp in female_prompts_for_dir]
    m_acts = [extract_activation(model, tokenizer, mp, layer_idx=layer) for mp in male_prompts_for_dir]
    f_mean = torch.stack(f_acts).mean(dim=0).squeeze()
    m_mean = torch.stack(m_acts).mean(dim=0).squeeze()
    g_dir = (f_mean - m_mean)
    g_dir = g_dir / g_dir.norm()
    
    # Extract activations for test prompts
    act_n = extract_activation(model, tokenizer, neutral_prompt, layer_idx=layer).squeeze()
    act_f = extract_activation(model, tokenizer, female_prompt, layer_idx=layer).squeeze()
    act_m = extract_activation(model, tokenizer, male_prompt, layer_idx=layer).squeeze()
    
    # Project onto gender direction
    proj_n = torch.dot(act_n.float(), g_dir.float()).item()
    proj_f = torch.dot(act_f.float(), g_dir.float()).item()
    proj_m = torch.dot(act_m.float(), g_dir.float()).item()
    
    closer = "FEMALE" if abs(proj_n - proj_f) < abs(proj_n - proj_m) else "MALE"
    
    print(f"  {layer:2d}  | {proj_n:10.2f}  | {proj_f:10.2f}  | {proj_m:10.2f}  | {closer}")
```

```python

# Excellent! Very clear mechanistic evidence:
# - From layer 10 onwards, the neutral prompt is consistently closer to the FEMALE direction
# - The gap grows substantially in deeper layers
# - At layer 40: neutral=9.70, female=36.79, male=-84.04 → neutral is MUCH closer to female

# Let's quantify: what fraction of the way from male to female is the neutral prompt?
print("Layer | Position on male-to-female axis (0=male, 1=female)")
print("-" * 60)

for layer in range(0, 42, 2):
    f_acts = [extract_activation(model, tokenizer, fp, layer_idx=layer) for fp in female_prompts_for_dir]
    m_acts = [extract_activation(model, tokenizer, mp, layer_idx=layer) for mp in male_prompts_for_dir]
    f_mean = torch.stack(f_acts).mean(dim=0).squeeze()
    m_mean = torch.stack(m_acts).mean(dim=0).squeeze()
    g_dir = (f_mean - m_mean)
    g_dir = g_dir / g_dir.norm()
    
    act_n = extract_activation(model, tokenizer, neutral_prompt, layer_idx=layer).squeeze()
    act_f = extract_activation(model, tokenizer, female_prompt, layer_idx=layer).squeeze()
    act_m = extract_activation(model, tokenizer, male_prompt, layer_idx=layer).squeeze()
    
    proj_n = torch.dot(act_n.float(), g_dir.float()).item()
    proj_f = torch.dot(act_f.float(), g_dir.float()).item()
    proj_m = torch.dot(act_m.float(), g_dir.float()).item()
    
    # Position: 0 = at male, 1 = at female
    if proj_f != proj_m:
        position = (proj_n - proj_m) / (proj_f - proj_m)
    else:
        position = 0.5
    
    bar = "█" * int(max(0, min(40, position * 40)))
    print(f"  {layer:2d}  | {position:.3f}  |{bar}")
```

```python

# One more validation: test across multiple different neutral prompts to make sure 
# this isn't specific to one prompt

neutral_prompts_varied = [
    [{"role": "user", "content": "What should I wear to a formal event?"}],
    [{"role": "user", "content": "Help me pick an outfit."}],
    [{"role": "user", "content": "What health screenings do I need?"}],
    [{"role": "user", "content": "Recommend a self-care routine."}],
    [{"role": "user", "content": "I need to buy a swimsuit."}],
    [{"role": "user", "content": "What deodorant should I use?"}],
    [{"role": "user", "content": "Tell me about myself."}],
    [{"role": "user", "content": "What vitamins should I take?"}],
    [{"role": "user", "content": "Help me plan my day."}],
    [{"role": "user", "content": "What should I cook for dinner?"}],
]

layer = 30  # Use a deep layer where signal is strong

# Compute gender direction at this layer
f_acts = [extract_activation(model, tokenizer, fp, layer_idx=layer) for fp in female_prompts_for_dir]
m_acts = [extract_activation(model, tokenizer, mp, layer_idx=layer) for mp in male_prompts_for_dir]
f_mean = torch.stack(f_acts).mean(dim=0).squeeze()
m_mean = torch.stack(m_acts).mean(dim=0).squeeze()
g_dir = (f_mean - m_mean)
g_dir = g_dir / g_dir.norm()

proj_f_ref = torch.dot(extract_activation(model, tokenizer, female_prompt, layer_idx=layer).squeeze().float(), g_dir.float()).item()
proj_m_ref = torch.dot(extract_activation(model, tokenizer, male_prompt, layer_idx=layer).squeeze().float(), g_dir.float()).item()

print(f"Reference: Female proj = {proj_f_ref:.2f}, Male proj = {proj_m_ref:.2f}")
print(f"\nNeutral prompt projections (0=male, 1=female):")
print("-" * 70)

positions = []
for np_msg in neutral_prompts_varied:
    act_n = extract_activation(model, tokenizer, np_msg, layer_idx=layer).squeeze()
    proj_n = torch.dot(act_n.float(), g_dir.float()).item()
    position = (proj_n - proj_m_ref) / (proj_f_ref - proj_m_ref)
    positions.append(position)
    bar = "█" * int(max(0, min(40, position * 40)))
    print(f"  {position:.3f}  |{bar}| {np_msg[0]['content'][:50]}")

print(f"\nMean position: {sum(positions)/len(positions):.3f}")
print(f"All closer to female: {all(p > 0.5 for p in positions)}")
```

```python

# Remarkable! Many neutral prompts are even MORE female than the explicit "I'm a woman" prompt!
# Mean position = 1.16 (beyond 1.0 = more female than explicitly stated female prompts)
# ALL 10/10 neutral prompts are closer to female than male.

# This confirms the finding very strongly. The model's internal representations 
# treat the user as female even more strongly than when the user explicitly says they're female.

# Let me do one final test: does this happen with a base model comparison?
# Let's check if PEFT adapters are responsible

print("Model info:")
print(f"  Type: {type(model).__name__}")
print(f"  Base model type: {type(model.base_model).__name__}")

# Check the PEFT config
if hasattr(model, 'peft_config'):
    for key, config in model.peft_config.items():
        print(f"\nPEFT config '{key}':")
        print(f"  Type: {config.peft_type}")
        print(f"  Target modules: {config.target_modules}")
        if hasattr(config, 'r'):
            print(f"  Rank: {config.r}")
        if hasattr(config, 'lora_alpha'):
            print(f"  Alpha: {config.lora_alpha}")
```
